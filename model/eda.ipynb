```python
# =========================================================
# 🔍 Exploratory Data Analysis for Genetic Disease Dataset
# =========================================================

# Install required packages if they are missing (works in Jupyter)
# %pip install -q pandas seaborn matplotlib

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# To display all columns
pd.set_option('display.max_columns', None)

# -----------------------
# 1. Load Dataset
# -----------------------
df = pd.read_csv("dataset.csv")

# -----------------------
# 2. Basic Dataset Overview
# -----------------------
print("📌 Dataset Shape:", df.shape)
print("\n📌 Column Names:\n", df.columns)
print("\n📌 First 5 Rows:\n")
display(df.head())

print("\n📌 Dataset Info:")
df.info()

print("\n📌 Summary Statistics:")
display(df.describe())

# -----------------------
# 3. Check Missing Values
# -----------------------
print("\n📌 Missing Values:")
print(df.isnull().sum())

plt.figure(figsize=(10,5))
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Value Heatmap")
plt.show()

# -----------------------
# 4. Disease Distribution
# -----------------------
plt.figure(figsize=(10,5))
sns.countplot(y='Disease_Status', data=df, palette="viridis")
plt.title("Distribution of Genetic Diseases")
plt.show()

# -----------------------
# 5. Numerical Feature Distribution
# -----------------------
numerical_cols = [
    "Age", "Maternal_Age_at_Birth", "Family_History_Score",
    "Hemoglobin_Level_g/dL", "MCV_fL", "MCH_pg"
]

df[numerical_cols].hist(figsize=(14,10), bins=30, color="skyblue")
plt.suptitle("Distribution of Numerical Features", fontsize=16)
plt.show()

# -----------------------
# 6. Correlation Heatmap
# -----------------------
plt.figure(figsize=(12,8))
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Between Numerical Features")
plt.show()

# -----------------------
# 7. Boxplots for Outlier Detection
# -----------------------
plt.figure(figsize=(14,10))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x=df[col], color="lightgreen")
    plt.title(f"{col} Boxplot")
plt.tight_layout()
plt.show()

# -----------------------
# 8. Relationship with Disease
# -----------------------
plt.figure(figsize=(14,6))
sns.boxplot(x='Disease_Status', y='Hemoglobin_Level_g/dL', data=df, palette="Set2")
plt.xticks(rotation=45)
plt.title("Hemoglobin Levels Across Diseases")
plt.show()

plt.figure(figsize=(14,6))
sns.boxplot(x='Disease_Status', y='MCV_fL', data=df, palette="Set1")
plt.xticks(rotation=45)
plt.title("MCV Levels Across Diseases")
plt.show()

# -----------------------
# 9. Countplots for Binary Indicators
# -----------------------
binary_cols = ["Is_Parent_Carrier", "Is_Sibling_Affected", "Has_Chronic_Cough", "Newborn_Screen_Flag"]

plt.figure(figsize=(12,8))
for i, col in enumerate(binary_cols, 1):
    plt.subplot(2, 2, i)
    sns.countplot(x=col, data=df, palette="magma")
    plt.title(f"Distribution of {col}")
plt.tight_layout()
plt.show()

# -----------------------
# 10. Disease Breakdown by Genetic Indicators
# -----------------------
for col in binary_cols:
    plt.figure(figsize=(10,5))
    sns.countplot(data=df, x=col, hue="Disease_Status", palette="tab20")
    plt.title(f"Disease Distribution by {col}")
    plt.show()

print("✅ EDA Completed Successfully!")
```
